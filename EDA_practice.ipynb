## Summary of Dataset ##
This dataset can be retrieved off of kaggle via [this link](https://www.kaggle.com/aimlanalytics/surgicalsiteinfectionsssishealthcare?select=ssi-sir-2017-ca-cdph-adult_28procs_.csv)

This dataframe was posted by the Center for Disease Control and Prevention National Healthcare Safety Network indicating the amount of surgical site infections in the year of 2017. This dataframe is composed of 19 columns that vary between objects and floats. This means that the data visualization process has to take into account categorical and numerical data. Columns 12 through 18 are missing thousands of values. This means that it would be rather unwise to create or try to utilize these columns when doing any predicitons. The last column is just Notes and contains a lot of missing values- since notes don't have much of an impact in the analysis process it would be wise to drop this row from the dataset. It would be best to list out missing values and see if anything can be done to replace them. 

## Checking and Editing Missing Values ##

```python
#first I am going to drop the notes column as the the amount of missing values is substancial and does not really impact the analysis 
df = df.drop('Notes', 1)
df.head(5)
```

```python
#another column I am going to delete is the year column as we already know that all of this data was retrieved in 2017
df = df.drop('Year', 1)
df.head(5)
```

```python
#Also dropping HAI and State as all of the data is retrieved from California and HAI is just listing that the infection came from a surgical healthcare setting
df = df.drop('HAI', 1)
df = df.drop('State', 1)
df.head(5)
```
